# Clase Cache

```python

from madridFines.cache import Cache, CacheError
cache = Cache("test_app", 10)

print("Directorio de cache:", cache.cache_dir)

# Crear archivo
cache.set("prueba.txt", "hola mundo")
print("Archivo creado:", cache.exists("prueba.txt"))

# Leer archivo
contenido = cache.load("prueba.txt")
print("Contenido leído:", contenido)

# Edad del archivo
edad = cache.how_old("prueba.txt")
print("Edad:", round(edad, 2))

# Eliminar un archivo
cache.delete("prueba.txt")
print("Eliminado:", not cache.exists("prueba.txt"))

# Excepcion cache
try:
    cache.load("no_existe.txt")
except CacheError as e:
    print("Error esperado:", e)

#limpiar cache
cache.clear()
```

# Clase CacheUrl

```python
from madridFines.cacheUrl import CacheUrl

url='https://books.toscrape.com/'
# Instancia
c = CacheUrl("test_app", 10)

contenido = c.get(url)
print("Contenido descargado:", contenido[:100])

# Verificar que ahora está en cache
print("Existe en cache:", c.exists(url))

#Leer desde cache sin pedir de nuevo
contenido_cache = c.load(url)
print("Contenido desde cache:", contenido_cache[:100])

# #Edad del archivo
edad = c.how_old(url)
print("Edad de url:", edad)

#Eliminarlo
c.delete(url)
print("Eliminado:", not c.exists(url))
```

# Clase MadridFines


```python
from madridFines.madridFines import MadridFines, MadridError

# Usar MadridFines para cargar y limpiar
fines = MadridFines("test_app", 10)
fines.add(2025, 4)
print("Meses y años cargados:", fines.loaded)

# Ver algunos análisis rápidos
print("Tabla fines por calificación:")
print(fines.fines_calification().head())

print("Total pago por mes/año:")
print(fines.total_payment().head())

#exportar la tabla por horas
fines.fines_hour('prueba.png')

#Prueba de la Exception
# Excepcion cache
try:
    fines.add(2030,4)
except MadridError as e:
    print(f"Año fuera de rango: 2030")

```
